```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MaxNLocator
import matplotlib.ticker as mtick
```

```python
test_data_round1 = pd.read_csv('../test_performance/test_round1.csv')
test_data_round2 = pd.read_csv('../test_performance/test_round2.csv')
test_data_round3 = pd.read_csv('../test_performance/test_round3.csv')
test_data_round4 = pd.read_csv('../test_performance/test_round4.csv')
test_data_round5 = pd.read_csv('../test_performance/test_round5.csv')
test_data_round6 = pd.read_csv('../test_performance/test_round6.csv')
test_data_round7 = pd.read_csv('../test_performance/test_round7.csv')
test_data_round8 = pd.read_csv('../test_performance/test_round8.csv')
```

```python
mae_sens = []
mae_emit = []
mae_rand = []
mae_large = []
mae_small = []
rmse_sens = []
rmse_emit = []
rmse_rand = []
rmse_large = []
rmse_small = []

dt = [test_data_round1, test_data_round2, test_data_round3, test_data_round4, test_data_round5, test_data_round6, test_data_round7, test_data_round8]
for i in range(8):
    mae_emit.append(dt[i]['S1_mae'][0] + dt[i]['T1_mae'][0])
    mae_sens.append(dt[i]['S1_mae'][3] + dt[i]['T1_mae'][3])
    mae_rand.append(dt[i]['S1_mae'][2] + dt[i]['T1_mae'][2])
    mae_large.append(dt[i]['S1_mae'][1] + dt[i]['T1_mae'][1])
    mae_small.append(dt[i]['S1_mae'][4] + dt[i]['T1_mae'][4])
    rmse_emit.append(dt[i]['S1_rmse'][0] + dt[i]['T1_rmse'][0])
    rmse_sens.append(dt[i]['S1_rmse'][3] + dt[i]['T1_rmse'][3])
    rmse_rand.append(dt[i]['S1_rmse'][2] + dt[i]['T1_rmse'][2])
    rmse_large.append(dt[i]['S1_rmse'][1] + dt[i]['T1_rmse'][1])
    rmse_small.append(dt[i]['S1_rmse'][4] + dt[i]['T1_rmse'][4])
```

```python
metric_mae_type1 = [mae_emit, mae_sens, mae_rand]
metric_mae_type2 = [mae_large, mae_small]
metric_rmse_type1 = [rmse_emit, rmse_sens, rmse_rand]
metric_rmse_type2 = [rmse_large, rmse_small]
step = range(1, 9)
```

```python
fig = plt.figure(num = 1, figsize = (6,6))

ax1 = fig.add_subplot(111)
ax1.plot(step, metric_mae_type2[0])
ax1.plot(step, metric_mae_type2[1])
#ax1.plot(step, metric_rmse_type1[2])
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
ax1.set_xlim(1, 8, 1)
ax1.set_xlabel('round')
ax1.set_ylabel('RMSE')
ax1.legend(['Large Mol', 'Small Mol'])
ax1.set_title('MAE of T1 and S1 per Round')

plt.savefig('metric_mae_type2.png')
plt.show()
```

```python
fig = plt.figure(num = 1, figsize = (6,6))

ax1 = fig.add_subplot(111)
ax1.plot(step, metric_mae_type1[0])
ax1.plot(step, metric_mae_type1[1])
ax1.plot(step, metric_mae_type1[2])
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
ax1.set_xlim(1, 8, 1)
ax1.set_xlabel('round')
ax1.set_ylabel('RMSE')
ax1.legend(['Emitters', 'Sensitizers', 'Random'])
ax1.set_title('MAE of T1 and S1 per Round')

plt.savefig('metric_mae_type1.png')
plt.show()
```

```python
fig = plt.figure(num = 1, figsize = (6,6))

ax1 = fig.add_subplot(111)
ax1.plot(step, metric_rmse_type2[0])
ax1.plot(step, metric_rmse_type2[1])
#ax1.plot(step, metric_rmse_type1[2])
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
ax1.set_xlim(1, 8, 1)
ax1.set_xlabel('round')
ax1.set_ylabel('RMSE')
ax1.legend(['Large Mol', 'Small Mol'])
ax1.set_title('RMSE of T1 and S1 per Round')

plt.savefig('metric_rmse_type2.png')
plt.show()
```

```python
fig = plt.figure(num = 1, figsize = (6,6))

ax1 = fig.add_subplot(111)
ax1.plot(step, metric_rmse_type2[0])
ax1.plot(step, metric_rmse_type2[1])
ax1.plot(step, metric_rmse_type1[2])
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
ax1.set_xlim(1, 8, 1)
ax1.set_xlabel('round')
ax1.set_ylabel('RMSE')
ax1.legend(['Emitters', 'Sensitizers', 'Random'])
ax1.set_title('RMSE of T1 and S1 per Round')

plt.savefig('metric_rmse_type1.png')
plt.show()
```
